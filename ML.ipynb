```python
import pandas as pd
```

```python
df=pd.read_csv('placement - placement.csv')
df
```

```python
df.isnull().sum()
```

```python
import matplotlib.pyplot as plt
plt.scatter(df["cgpa"],df["package"])
plt.xlabel("CGPA")
plt.ylabel("Package")
```

```python
x=df[['cgpa']]#independent feature
y=df[['package']]#dependent feature
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=39)
```

```python
x_train.shape,y_train.shape
```

```python
x_train
```

```python
y_train
```

```python
x_test.shape,y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
print(dir(model))
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)*100
```

```python
model.predict([[8.60]]),model.predict([[5.38]])
```

```python
model.predict([[9.0]])
```

```python
model.predict([[9.1]])
```

```python
#y_test
```

```python
y_pred_test=model.predict(x_test)
y_pred_test
```

```python
y_test
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred_test)
```
